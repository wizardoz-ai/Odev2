```python
# Öncelikle kullanacağımız modellerimizi çağıralım.

from ultralytics import YOLO
import os
import cv2
import matplotlib.pyplot as plt
```

```python
# YOLO11x modelini yükleyelim.

model = YOLO('yolo11x.pt')
```

```python
# Fotoğraf ve video dosyalarının yollarını tanımlayalım.

image_folder = 'images'  # Fotoğrafların olduğu klasör
video_file = 'images/video1.mp4'  # Video dosyası

# Tüm fotoğraf dosyalarını listeleyelim.
image_files = [os.path.join(image_folder, file) for file in os.listdir(image_folder) if file.endswith(('.jpeg', '.png'))]
print(f"Yüklenecek fotoğraflar: {image_files}")
print(f"Yüklenecek video: {video_file}")
```

```python
#Dosyalarımızın yolunu belirledikten sonra modelimizi çalıştırabilir ve sonuçlarımızı görebiliriz.

source = "images" 

results = model(source, stream = False, conf = 0.45, show = True, save = True, device = 'mps')

print(results)
```
